```python

import tkinter as tk
from tkinter import messagebox


def calculate_orientation():
    try:
        # Get marks from inputs
        math = float(entry_math.get())
        physics = float(entry_physics.get())
        arabic = float(entry_arabic.get())
        french = float(entry_french.get())

        student_marks = {
            "Mathematics": math,
            "Physics": physics,
            "Arabic": arabic,
            "French": french
        }

        # Orientation coefficients (example)
        orientations = {
            "Science Mathematics": {
                "Mathematics": 9,
                "Physics": 7,
                "Arabic": 2
            },
            "Science Physics": {
                "Physics": 7,
                "Mathematics": 7,
                "Arabic": 2
            }
        }

        results = {}

        for orientation, subjects in orientations.items():
            score = 0
            coef_sum = 0

            for subject, coef in subjects.items():
                score += student_marks.get(subject, 0) * coef
                coef_sum += coef

            percentage = (score / (coef_sum * 20)) * 100
            results[orientation] = percentage

        # Sort orientations by percentage
        sorted_results = sorted(
            results.items(),
            key=lambda x: x[1],
            reverse=True
        )

        # Display result
        result_text = "Orientation Ranking:\n\n"
        for orientation, percent in sorted_results:
            result_text += f"{orientation}: {percent:.2f}%\n"

        label_result.config(text=result_text)

    except ValueError:
        messagebox.showerror("Input Error", "Please enter valid numbers (0â€“20).")


# Create main window
root = tk.Tk()
root.title("Student Orientation App")
root.geometry("350x400")

# Title
tk.Label(root, text="Student Orientation App", font=("Arial", 16)).pack(pady=10)

# Inputs
tk.Label(root, text="Mathematics").pack()
entry_math = tk.Entry(root)
entry_math.pack()

tk.Label(root, text="Physics").pack()
entry_physics = tk.Entry(root)
entry_physics.pack()

tk.Label(root, text="Arabic").pack()
entry_arabic = tk.Entry(root)
entry_arabic.pack()

tk.Label(root, text="French").pack()
entry_french = tk.Entry(root)
entry_french.pack()

# Button
tk.Button(
    root,
    text="Calculate Orientation",
    command=calculate_orientation
).pack(pady=15)

# Result label
label_result = tk.Label(root, text="", font=("Arial", 11), justify="left")
label_result.pack()

# Run app
root.mainloop()
```
